```python
import pandas as pd
```

```python
df = pd.read_csv('genomes.insect.for_graph.230705.tab', sep='\t', header=None)
df.columns = ['TaxonID', 'Species', 'sp', 'Genus', 'Family', 'Order', 'Phylum', 'GenomeSize']
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x= 'Order', data = df, order = df['Order'].value_counts().index, log=True)
plt.xticks(rotation=90)
```

```python
df_grouped = df.groupby('Order')
df_grouped.mean('GenomeSize').plot.bar(y='GenomeSize')
```

```python
fig, ax= plt.subplots(figsize=(18, 6))
sns.boxenplot(x= 'Order', y='GenomeSize', data = df)
ax.set_yscale('log')
plt.xticks(rotation=90)
plt.show()
```
